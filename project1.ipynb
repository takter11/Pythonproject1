```
import pandas as pd
import numpy as np
import seaborn as sns
% matplotlib inline
missing_values=['na','s']
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv',na_values=missing_values)
df.head(100)
dfJHS=df[(df['School DBN']=='01M034') & (df['Regents Exam']=='Living Environment')]
dfJHS1=dfJHS[['Year','Mean Score','Regents Exam']]
dfJHS1.drop_duplicates(keep='first',inplace=True)
dfJHS1
dfJHSFDR=dfJHS1.groupby(['Year'])['Mean Score'].mean()

dfjhs=df[(df['School DBN']=='84X717') & (df['Regents Exam']=='Living Environment')]
dfjhs1=dfjhs[['Year','Mean Score','Regents Exam']]
dfjhs1.drop_duplicates(keep='first',inplace=True)
dfjhs1
dfjhsICS=dfjhs1.groupby(['Year'])['Mean Score'].mean()
#comparison plot
left=dfJHSFDR
right=dfjhsICS
comp=pd.merge(left,right, on='Year')
comp=comp.rename(columns={'Mean Score_x':'dfJHSFDR','Mean Score_y': 'dfjhsICS'})
comp.plot(kind='bar')
comp.plot(kind='hist')
comp.plot(kind='line')
# My interpretation 
#In the year 2015 both schools mean score are almost same (dfJHSFDR=78.04 and dfjhsICS= 78.08) 
#and in the year 2016, dfJHSFDR has less mean score(73.75) than dfjhsICS (77.10)
# Overall in 2016 both schools mean score then than in the year 2016
#dfJHSFDR= P.S. 034 Franklin D. Roosevelt	
#dfjhsICS=Icahn Charter School
```
